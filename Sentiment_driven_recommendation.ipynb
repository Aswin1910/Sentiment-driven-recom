```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset1 = pd.read_csv('appliance_reviews.csv')
```

```
import gzip
import json

def parse(path):
    with gzip.open(path, 'rt', encoding='utf-8') as g:
        for l in g:
            yield json.loads(l)

def get_df(path, max_rows=5000):
    data = []
    for i, d in enumerate(parse(path)):
        if i >= max_rows:
            break
        data.append(d)
    return pd.DataFrame(data)

df = get_df("Appliances.jsonl.gz")
df[['rating', 'title', 'text', 'user_id','asin']].dropna().to_csv("dataset.csv", index=False)
```

```
import numpy as np
dataset = pd.read_csv('dataset.csv')
dataset.head()
```

```
#Train


import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus1 = []
for i in range(0, 5000):
  review = re.sub('[^a-zA-Z]', ' ', dataset1['review_text'][i])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  all_stopwords.remove("didn't")
  all_stopwords.remove("doesn't")
  all_stopwords.remove("don't")
  all_stopwords.remove("hasn't")
  all_stopwords.remove("haven't")
  all_stopwords.remove("isn't")
  all_stopwords.remove("wasn't")
  all_stopwords.remove("weren't")
  all_stopwords.remove("won't")
  all_stopwords.remove("wouldn't")
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus1.append(review)
```

```
#APPLIANCES


import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 5000):
  if isinstance(dataset['text'][i], str):
    review = re.sub('[^a-zA-Z]', ' ', dataset['text'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    all_stopwords.remove("didn't")
    all_stopwords.remove("doesn't")
    all_stopwords.remove("don't")
    all_stopwords.remove("hasn't")
    all_stopwords.remove("haven't")
    all_stopwords.remove("isn't")
    all_stopwords.remove("wasn't")
    all_stopwords.remove("weren't")
    all_stopwords.remove("won't")
    all_stopwords.remove("wouldn't")
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
  else:
    corpus.append('')
```
